```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/User/Desktop/nasashit/BloomAI/V2/training/training_dataset.csv')
df_clean = df.dropna(subset=['EVI']).copy()
# 1. INVESTIGATE THE TEMPORAL PATTERN
print("=== INVESTIGATING TEMPORAL BIAS ===")
bloom_by_month = df_clean.groupby('month')['bloom'].mean()
print("Bloom frequency by month:")
print(bloom_by_month.sort_values(ascending=False))

# 2. CHECK IF BLOOM IS JUST SEASONAL
plt.figure(figsize=(10, 6))
bloom_by_doy = df_clean.groupby('day_of_year')['bloom'].mean()
plt.plot(bloom_by_doy.index, bloom_by_doy.values)
plt.title('Bloom Frequency by Day of Year')
plt.xlabel('Day of Year')
plt.ylabel('Bloom Frequency')
plt.show()
```
